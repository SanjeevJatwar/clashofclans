### **Drop sum more column :-**

```python
FINAL_DROP = [

    
    # Duplicates — perfectly correlated with columns you already have
    "avg_balance_y",          # duplicate of avg_balance_x
    "quarterly_avg_balance",  # correlated with monthly_avg_balance
    "daily_avg_balance",      # correlated with avg_balance_x
    "credit_ratio",           # = 1 - debit_ratio
    "debit_ratio",            # keep credit_debit_ratio instead
    "day_txn_rate",           # = 1 - night_txn_rate
    "avg_txn_per_week",       # correlated with avg_txn_per_day
    "avg_txn_per_month",      # correlated with avg_txn_per_day
    "max_txn_1week",          # correlated with max_txn_1day
    
    # Admin — branch operational metrics, not account behaviour
    "branch_employee_count",
    "branch_turnover",
    "branch_asset_size",
    "branch_type",
    "scheme_code",
    "currency_code",          # near zero variance — almost all INR
    "num_chequebooks",        # cheque_availed already captures this
    
    # High cardinality strings still remaining
    "most_common_mcc",        # needs frequency encoding or drop
    "least_common_mcc_code",  # needs frequency encoding or drop
]

df = df.drop(columns=[c for c in FINAL_DROP if c in df.columns])
print(f"Dropped : {len([c for c in FINAL_DROP if c in df.columns])} columns")
print(f"Remaining : {df.shape[1]} columns")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import (classification_report, roc_auc_score,
                             average_precision_score, confusion_matrix,
                             f1_score, precision_recall_curve)
import lightgbm as lgb
import xgboost as xgb
from imblearn.combine import SMOTETomek
import warnings
warnings.filterwarnings("ignore")
```

```python
X = df_train.drop(columns=["is_mule","account_id"],axis = 1)
y = df_train["is_mule"]

print(f"Features : {X.shape[1]}")
print(f"Mule rate: {y.mean():.4f}")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
print(f"\nTrain shape : {X_train.shape}")
print(f"Test shape  : {X_test.shape}")
print(f"Train mules : {y_train.sum()} ({y_train.mean():.4f})")
print(f"Test mules  : {y_test.sum()} ({y_test.mean():.4f})")
```

```python
# ── LightGBM — best model for this type of data ─────────────
# scale_pos_weight not needed after SMOTETomek
# but class_weight='balanced' adds extra safety
lgbm = lgb.LGBMClassifier(
    n_estimators      = 1000,
    learning_rate     = 0.05,
    num_leaves        = 63,
    max_depth         = -1,
    min_child_samples = 20,
    subsample         = 0.8,
    colsample_bytree  = 0.8,
    reg_alpha         = 0.1,
    reg_lambda        = 0.1,
    class_weight      = "balanced",
    random_state      = 42,
    n_jobs            = -1,
    verbose           = -1
)

lgbm.fit(
    X_train, y_train,
    eval_set=[(X_test, y_test)],
    callbacks=[lgb.early_stopping(50, verbose=False),
               lgb.log_evaluation(100)]
)

# ── Evaluate ─────────────────────────────────────────────────
y_pred  = lgbm.predict(X_test)
y_proba = lgbm.predict_proba(X_test)[:, 1]

print("\n" + "="*50)
print("LIGHTGBM RESULTS")
print("="*50)
print(f"ROC-AUC  : {roc_auc_score(y_test, y_proba):.4f}")
print(f"PR-AUC   : {average_precision_score(y_test, y_proba):.4f}")
print(f"F1-Mule  : {f1_score(y_test, y_pred):.4f}")
print(f"F1-Macro : {f1_score(y_test, y_pred, average='macro'):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred,
      target_names=["legit", "mule"]))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
