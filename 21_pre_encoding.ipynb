```python
CHANNEL_GROUPS = {
    # Digital payments — UPI, IMPS, NEFT, online
    "digital":    ["UPC", "UPD", "IPM", "NTD", "END", "OPI", "PCA", "MCR", "MAC"],
    
    # Fund transfers — internal and external
    "transfer":   ["FTD", "FTC", "P2A", "IFD", "IFC", "ETD", "CTC", "SID", "IAD"],
    
    # Auto / standing instructions — scheduled payments
    "auto":       ["STD", "STC", "APD", "TPD", "TPC", "MAD", "RCD"],
    
    # Cash — physical cash movement (strong mule signal)
    "cash":       ["ATW", "CSD", "OCD", "NWD", "SCW", "CHQ", "CHD"],
    
    # Return / clearing — reversals and clearing house
    "return":     ["RTD", "CCL", "ASD"],
    
    # No transaction
    "no_txn":     ["NO_TXN"]
}

# Reverse map — code → group
code_to_group = {}
for group, codes in CHANNEL_GROUPS.items():
    for code in codes:
        code_to_group[code] = group

print(code_to_group)
```

```python
for col in ["least_common_channel", "most_common_channel"]:
    if col in df1.columns:
        df1[f"{col}_group"] = df1[col].map(code_to_group).fillna("other")
        print(f"\n{col}_group distribution:")
        print(df1[f"{col}_group"].value_counts())
```

```python
# Map by risk level — higher number = higher mule/fraud association
CHANNEL_GROUP_MAP = {
    "digital":  0,   # normal everyday usage
    "auto":     1,   # scheduled — low risk
    "transfer": 2,   # fund movements — medium risk
    "return":   3,   # reversals/chargebacks — elevated risk
    "cash":     4,   # physical cash — highest mule signal
    "no_txn":   5,   # never used this channel — dormant signal
    "other":    2,   # unknown codes — treat as transfer level
}

for col in ["least_common_channel_group", "most_common_channel_group"]:
    if col in df1.columns:
        df1[col] = df1[col].map(CHANNEL_GROUP_MAP)
        print(f"\n{col} distribution after mapping:")
        print(df1[col].value_counts().sort_index())
        print(f"Nulls: {df1[col].isna().sum()}")
        
df1 = df1.drop(columns=["least_common_channel", "most_common_channel"], 
               errors="ignore")
```

```python
# Normalise order first — CDM,CRM and CRM,CDM are identical
df1["atm_deposit_types"] = df1["atm_deposit_types"].replace({
    "CRM,CDM": "CDM,CRM",   # standardise to one form
    "REC":     "CRM"         # recycler = CRM, merge tiny group
})

# Now create binary features — each machine type is its own signal
df1["uses_cdm"] = df1["atm_deposit_types"].str.contains("CDM", na=False).astype(int)
df1["uses_crm"] = df1["atm_deposit_types"].str.contains("CRM", na=False).astype(int)

# Derived features
df1["uses_both_atm"]  = ((df1["uses_cdm"] == 1) & (df1["uses_crm"] == 1)).astype(int)
df1["no_atm_deposit"] = (df1["atm_deposit_types"] == "NO_ATM").astype(int)

# Drop the original string column — no longer needed
df1 = df1.drop(columns=["atm_deposit_types"])

# Verify
print(df1[["uses_cdm","uses_crm","uses_both_atm","no_atm_deposit"]].value_counts())
```
