```python
import pandas as pd
```

```python
df = pd.read_parquet("branch.parquet")
print(df.shape)
df.head()
```

```python
df1 = pd.read_parquet("accounts.parquet")
print(df1.shape)
df1.head()
```

```python
df_final = df1.merge(df, on="branch_code", how="left")
print(df_final.shape)
df_final.sample(5)
```

```python
df3 = pd.read_parquet("accounts-additional.parquet")
print(df3.shape)
df3.head()
```

```python
df3.account_id.nunique()
```

```python
df4 = pd.read_parquet("customer_account_linkage.parquet")
print(df4.shape)
df4.head()
```

```python
df4.customer_id.nunique()
```

```python
df4.account_id.nunique()
```

### Merge account columns : -

```python
df_result = df_final.merge(df3, on="account_id", how="left")
print(df_result.shape)
df_result.sample(5)
```

```python
df_result_f = df_result.merge(df4, on="account_id", how="left")
print(df_result_f.shape)
df_result_f.sample(5)
```

```python
df_result_f.to_parquet("f_a.parquet", engine="pyarrow", compression="snappy")
```
