```python
df1 = df_main
```

```python
no_txn_mask = df1["total_transactions"] == 0
no_txn_mask.sum()
```

```python
df1["avg_balance_missing"] = df1["avg_balance_x"].isna().astype(int)

new_accounts = df1["days_since_account_open"] < 30
df1.loc[new_accounts & df1["avg_balance_x"].isna(), "avg_balance_x"] = 0

#fill with median WITHIN the same product_family
# because savings accounts and current accounts have very different balance profiles
df1["avg_balance_x"] = df1.groupby("product_family")["avg_balance_x"].transform(
    
    lambda x: x.fillna(x.median())
)
print(df1.shape)
df1.sample(3)
```

```python
df1.isnull().sum()  
```

```python
df3 = df1[["account_id","account_opening_date_y","account_opening_date_x"]]
print(df3.shape)
df3.sample(3)
```

```python
df2 = df1
df2['active_tenure_days'] = (
    (pd.to_datetime(df2['account_opening_date_y']) - 
     pd.to_datetime(df2['relationship_start_date']))
    .dt.days
).fillna(0)
print(df2.shape)
df2.sample(3)
```

```python
df2.isnull().sum()
```

```python
df3 = df2
```

```python
# 1️⃣ Create a flag for missing monthly average balance
df3["monthly_avg_balance_missing"] = df3["monthly_avg_balance"].isna().astype(int)

# 2️⃣ If the account is very new (<30 days) and balance is missing → set to 0
new_accounts = df3["days_since_account_open"] < 30
df3.loc[new_accounts & df3["monthly_avg_balance"].isna(), "monthly_avg_balance"] = 0

# 3️⃣ Fill remaining missing values using median within the same product_family
df3["monthly_avg_balance"] = df3.groupby("product_family")["monthly_avg_balance"].transform(
    lambda x: x.fillna(x.median())
)

print(df3.shape)
df3.sample(3)
```

```python
# Find every object column that still has float values hiding inside
problem_cols = []
for col in df9.select_dtypes(include="object").columns:
    has_float = df9[col].apply(lambda x: isinstance(x, float)).any()
    if has_float:
        problem_cols.append(col)

print("Columns with mixed float/string:", problem_cols)
```

```python
# These two still have float NaN hiding inside despite fillna
for col in ['most_common_mcc', 'least_common_mcc_code']:
    df9[col] = df9[col].astype(str)
    df9[col] = df9[col].replace({"nan": "NO_TXN", "None": "NO_TXN"})

# Verify
for col in ['most_common_mcc', 'least_common_mcc_code']:
    has_float = df9[col].apply(lambda x: isinstance(x, float)).any()
    print(f"{col} — float remaining: {has_float}")
    print(f"  unique sample: {df9[col].unique()[:5]}")
```

```python
def clean_mcc_code(val):
    if val == "NO_TXN":
        return val
    try:
        return str(int(float(val)))  # "9384.0" → 9384 → "9384"
    except:
        return "NO_TXN"

df9["most_common_mcc"]      = df9["most_common_mcc"].apply(clean_mcc_code)
df9["least_common_mcc_code"] = df9["least_common_mcc_code"].apply(clean_mcc_code)

# Verify
print(df9["most_common_mcc"].unique()[:5])
# Should show: ['NO_TXN' '9384' '5411' '2017' '5682']
```
